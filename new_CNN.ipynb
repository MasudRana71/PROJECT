```python
import os
import cv2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Set the path to your image directory
image_directory = "/content/Monkeypox_dataset_22"

# Set the image dimensions
image_size = (128,128)

# Load and preprocess the images
images = []
labels = []

for class_label in os.listdir(image_directory):
    class_dir = os.path.join(image_directory, class_label)
    for image_file in os.listdir(class_dir):
        image_path = os.path.join(class_dir, image_file)
        image = cv2.imread(image_path)
        image = cv2.resize(image, image_size)
        images.append(image)
        labels.append(class_label)


```

```python
# Convert the lists to arrays
images = np.array(images)
labels = np.array(labels)

# Encode the class labels as integers
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)

# Normalize the image data
images = images / 255.0
```

```python
# Perform one-hot encoding on the labels
num_classes = len(np.unique(labels))
labels = tf.keras.utils.to_categorical(labels, num_classes)


```

```python
print("Images array shape:", images.shape)
print("Labels array shape:", labels.shape)
```

```python
# Display the array output of images
for i in range(len(images)):
    print(f"Image {i+1}:")
    print(images[i])
    print()
```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.3, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128,128, 3)),

    Dropout(0.25),  # Dropout added after the first MaxPooling2D layer

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Dropout(0.25),  # Dropout added after the second MaxPooling2D layer

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Conv2D(256, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),


    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.25),  # Dropout added after the first Dense layer
    Dense(num_classes, activation='softmax')
])
```

```python
from tensorflow.keras.optimizers import Adam

learning_rate = 0.0003


# Compile the model with a custom learning rate
model.compile(optimizer=Adam(learning_rate=learning_rate),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=30, validation_data=(X_val, y_val))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Make predictions on the validation set
y_pred = model.predict(X_val)
y_pred_labels = np.argmax(y_pred, axis=1)
y_val_labels = np.argmax(y_val, axis=1)

# Compute the confusion matrix
cm = confusion_matrix(y_val_labels, y_pred_labels)

# Define class labels
class_labels = label_encoder.classes_

# Plot the confusion matrix with color and labels
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens", xticklabels=class_labels, yticklabels=class_labels)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Make predictions on the validation set
y_pred = model.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_val, axis=1)

# Generate confusion matrix
cm = confusion_matrix(y_true_classes, y_pred_classes)
print("Confusion Matrix:")
print(cm)

# Generate classification report
target_names = label_encoder.classes_
report = classification_report(y_true_classes, y_pred_classes, target_names=target_names)
print("Classification Report:")
print(report)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_val, y_val)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

# Binarize the labels for ROC curve computation
y_val_bin = label_binarize(y_val, classes=np.arange(num_classes))

# Compute the predicted probabilities for each class
y_pred_prob = model.predict(X_val)

# Compute the false positive rate (FPR), true positive rate (TPR), and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_val_bin[:, i], y_pred_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC AUC
fpr["micro"], tpr["micro"], _ = roc_curve(y_val_bin.ravel(), y_pred_prob.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot the ROC curve for each class and the micro-average ROC curve
plt.figure()
colors = ['blue', 'red', 'green', 'orange']  # Specify colors for each class
for i, color in zip(range(num_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label='Class {} (AUC = {:.2f})'.format(i, roc_auc[i]))

plt.plot(fpr["micro"], tpr["micro"], color='deeppink', lw=2, linestyle='--', label='Micro-average (AUC = {:.2f})'.format(roc_auc["micro"]))
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')  # Plot the diagonal line

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
